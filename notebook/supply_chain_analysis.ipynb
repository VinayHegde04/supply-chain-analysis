```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/supply_chain_data.csv")
df.head()
```

```python
df[['Product type','Number of products sold']].groupby('Product type').sum().sort_values(by='Number of products sold',ascending=False)
```

```python
product_sales = df.groupby('Product type')['Number of products sold'].sum().sort_values(ascending=False)
product_sales
```

```python
plt.Figure()
product_sales.plot(kind='bar')
plt.title('Total products sold by product type')
plt.xlabel('Product type')
plt.ylabel('Number of products sold')
plt.show()
```

```python
revenue_by_product= df.groupby('Product type')['Revenue generated'].sum().sort_values(ascending=False)
revenue_by_product
```

```python
plt.figure()
product_sales.plot(kind='bar')
plt.title("Revenue Generated by Product Type")
plt.xlabel('Product type')
plt.ylabel('Revenue generated')
plt.show
```

```python
Supplier_sales= df.groupby('Supplier name')['Number of products sold'].sum().sort_values(ascending=False)
Supplier_sales
```

```python
plt.figure()
Supplier_sales.plot(kind='bar')
plt.title("Supplier Performance Based on Products Sold")
plt.xlabel("Supplier")
plt.ylabel("Products Sold")
plt.show()
```

```python
Shipping_time= df.groupby('Product type')['Shipping times'].mean().sort_values(ascending=False)
Shipping_time
```

```python
plt.figure()
Shipping_time.plot(kind='bar')
plt.title("Average Shipping Time by Product Type")
plt.xlabel("Product Type")
plt.ylabel("Average Shipping Time (Days)")
plt.show()
```

```python
stock_levels = df.groupby('Product type')['Stock levels'].mean().sort_values(ascending=False)
stock_levels
```

```python
plt.figure()
stock_levels.plot(kind='bar')
plt.title("Average Stock Levels by Product Type")
plt.xlabel("Product Type")
plt.ylabel("Average Stock Level")
plt.show()
```

## Key Business Insights

1. Skincare products have the highest demand and generate the most revenue, making them the most profitable product category.

2. Cosmetics products have the lowest sales and revenue but maintain the highest stock levels, indicating possible overstocking.

3. Supplier 1 contributes the highest number of product sales, while Supplier 4 contributes the least.

4. Cosmetics products experience the longest shipping times, suggesting possible logistics inefficiencies.

5. Skincare products have the lowest shipping time but strong demand, indicating efficient supply chain handling for this category.

6. Inventory allocation may need adjustment, as cosmetics appear overstocked while skincare products may require higher stock levels.
